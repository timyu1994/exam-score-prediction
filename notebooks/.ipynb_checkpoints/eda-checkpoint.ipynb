### Data Transformation

```python
# Drop rows with null values for target variable
df.dropna(axis=0, subset=['final_test'], inplace=True)
```

```python
# Examine reasons for data duplication
df.loc[df.duplicated('student_id', keep=False)].sort_values(by=['student_id']).head(20)
```

```python
# Drop duplicates, keeping the first occurence
df.drop_duplicates('student_id', inplace=True)
```

```python
# Check that duplicates have been removed
df.shape
```

```python
# Combine sleep_time and wake_time into single feature
df[['sleep_time', 'wake_time']] = df[['sleep_time', 'wake_time']].astype('datetime64[ns]')
df['sleep_hours'] = df['wake_time'] - df['sleep_time']
df['sleep_hours'] = df['sleep_hours'].dt.components['hours'] + df['sleep_hours'].dt.components['minutes']
```

```python
# Check values for hours slept
print(f"sleep_hours : {np.unique(df['sleep_hours'].values)}")
```

```python
df.drop(df[['sleep_time', 'wake_time']], axis=1, inplace=True)
```

```python
# Standardize 'CCA' values to lowercase
df['CCA'] = [x.lower() for x in df['CCA']]
```

```python
# Standardize 'tuition' values to Y or N
df['tuition'] = [x[0].lower() for x in df['tuition']]
```

```python
# Check negative values for 'age'
df.loc[df['age'] < 0]
```

```python
# Since there's only 5 rows of negative values compared to the total 14,559 rows, the loss of information is not significant. To drop all instead of assuming the original values
df.drop(df[df['age'] < 0].index, inplace=True)
```

```python
# To replace ages 5 and 6 with the assumption that they refer to 15 and 16 respectively
df['age'] = [x + 10 if x < 7 else x for x in df['age'] ]
```

```python

for features, values in df.nunique().items():
    if(values < 14):
        print(f"{features} : {np.unique(df[features].values)}")
```

### Data Visualization

```python
# Data visualization using graphs for numerical data
sns.pairplot(df)
```

```python
# Data visualization using correlation heatmap for numerical data
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
cmap = sns.diverging_palette(h_neg=10, h_pos=240, as_cmap=True)
sns.heatmap(df.corr(), mask=mask, center=0, cmap=cmap, linewidths=1, annot=True, fmt='.2f')
```

```python
# Data visualization using boxplot for binary and categorical data
fig, axes = plt.subplots(4, 2, figsize=(18, 35))

bag_colors = {"yellow": "yellow", "green": "green", "white":"white", "red":"red", "blue":"blue", "black":"black"}
sns.boxplot(ax=axes[0, 0], y='final_test', x='bag_color', hue='bag_color', data=df, palette=bag_colors).set_title('Bag color vs Test Scores', weight='bold', fontsize=20)
axes[0,0].set_xlabel('Bag Color', fontsize=15)
axes[0,0].set_ylabel('Test Scores', fontsize=15)
axes[0,0].get_legend().remove()

sns.boxplot(ax=axes[0, 1], y='final_test', x='CCA', data=df, palette='viridis').set_title('CCA vs Test Scores', weight='bold', fontsize=20)
axes[0,1].set_xlabel('CCA', fontsize=15)
axes[0,1].set_ylabel('Test Scores', fontsize=15)

sns.boxplot(ax=axes[1, 0], y='final_test', x='mode_of_transport', data=df, palette='viridis').set_title('Mode of Transport vs Test Scores', weight='bold', fontsize=20)
axes[1,0].set_xlabel('Mode of Transport', fontsize=15)
axes[1,0].set_ylabel('Test Scores', fontsize=15)

sns.boxplot(ax=axes[1, 1], y='final_test', x='direct_admission', data=df, palette='viridis').set_title('Direct Admission vs Test Scores', weight='bold', fontsize=20)
axes[1,1].set_xlabel('Direct Admission', fontsize=15)
axes[1,1].set_ylabel('Test Scores', fontsize=15)

sns.boxplot(ax=axes[2, 0], y='final_test', x='learning_style', data=df, palette='viridis').set_title('Learning Style vs Test Scores', weight='bold', fontsize=20)
axes[2,0].set_xlabel('Learning Style', fontsize=15)
axes[2,0].set_ylabel('Test Scores', fontsize=15)

sns.boxplot(ax=axes[2, 1], y='final_test', x='gender', data=df, palette='viridis').set_title('Gender vs Test Scores', weight='bold', fontsize=20)
axes[2,1].set_xlabel('Gender', fontsize=15)
axes[2,1].set_ylabel('Test Scores', fontsize=15)

sns.boxplot(ax=axes[3, 0], y='final_test', x='tuition', data=df, palette='viridis').set_title('Tuition vs Test Scores', weight='bold', fontsize=20)
axes[3,0].set_xlabel('Tuition', fontsize=15)
axes[3,0].set_ylabel('Test Scores', fontsize=15)

fig.delaxes(axes[3][1])
```

<b><u>Conclusion</b></u>
<br>
Based on the graphs and charts displayed, the following inferences can be made from the data:

<u>Numerical</u>
1. Students who have no siblings tend to have lower attendance rates as well as lower overall scores compared to having 1 or 2. However, while students who have 1 sibling had similar attendance rates to those with 2 siblings, those with 2 siblings never scored above 90
2. No. of females in a class is negatively correlated with the no. of males which makes sense since class sizes are limited. Correlation is not perfect however due to different class sizes.
3. Most students with less than 5 tuition hours could not score higher than 50 but those that could scored 80. Students with tuition hours between 5 and 10 produce the highest possible scores. Lastly students with more than 10 tuition hours could not score above 80, and those above 15 tuition hours will not have a failing score.
4. Attendance rates below 90 and sleep hours below 7 guarantees that the student will have a failing score and is highly correlated with sleep hours. To consider dropping either attendance rates or sleep hours due to high correlation to improve model performance.
5. Age has no effect on finals score and can be dropped during model tuning.

<u>Categorical/Binary</u>
1. Bag colors, mode of transport and gender have little to no effect on final scores and can be dropped to improve model performace as they will likely generate noise for the model.
2. Students who have no CCA are more likely to score better than those with. However it is not guaranteed that a student with no CCA will have high scores.
3. Students who had direct admission are more likely to score better than those without as well. However it is not guaranteed that a student with direct admission will have high scores.
4. Students with the visual learning style are more likely to score better than those with the auditory learning style.
5. Students with tuition are more likely to score better than those without.

Generally, it can be observed that the relationship between features and the final scores tend to be non-linear as most numerical features are not highly correlated with the final scores. Even if there appears to be a relationship based on the seaborn plots (such as no. of siblings, tuitions hours per week, attendance rate, and sleep hours), the relationship is not continuous. There are also multiple binary and categorical features making noticeable differences in final scores.
